## Handling Missing Values in Training Data

```python
count=dataset_train.apply(lambda x: x.count(), axis=1).values
x_train=[]
for i in range(len(count)):
    if count[i]<3:
        continue
    else:
        x_train.append(dataset_train.iloc[i, :].values)
dataset_train=pd.DataFrame(x_train)
```

```python
dataset_train=dataset_train=dataset_train.replace(np.nan, " ")
dataset_train[0]=dataset_train[0].astype(str)
dataset_train[1]=dataset_train[1].astype(str)
dataset_train[2]=dataset_train[2].astype(str)
dataset_train[3]=dataset_train[3].astype(str)
dataset_train[4]=dataset_train[4].astype("int64")
```

## Separating Features and Target

```python
#no. of training data points
n=len(dataset_train)
X_train=dataset_train.iloc[:n, :-1].values
y_train=dataset_train.iloc[:n, -1].values
len(X_train)
```

## Importing Testing Data

```python
dataset_test=pd.read_csv("../input/amazon-ml-challenge-2021-hackerearth/test.csv", escapechar = "\\", quoting = csv.QUOTE_NONE)
dataset_test=dataset_test.replace(np.nan, " ")
dataset_test["TITLE"]=dataset_test["TITLE"].astype(str)
dataset_test["DESCRIPTION"]=dataset_test["DESCRIPTION"].astype(str)
dataset_test["BULLET_POINTS"]=dataset_test["BULLET_POINTS"].astype(str)
dataset_test["BRAND"]=dataset_test["BRAND"].astype(str)
dataset_test.describe()
```

```python
X_test=dataset_test.iloc[:, :].values
len(X_test)
```

## Cleaning the Texts

```python
for i in range(len(X_train)):
    X_train[i, 0]=X_train[i, 0]+" "+X_train[i, 1]+" "+X_train[i, 2]
    X_train[i, 3]=X_train[i, 3].lower()
X_train=np.delete(X_train, 1, 1)
X_train=np.delete(X_train, 1, 1)

for i in range(len(X_test)):
    X_test[i, 1]=X_test[i, 1]+" "+X_test[i, 2]+" "+X_test[i, 3]
    X_test[i, 4]=X_test[i, 4].lower()
X_test=np.delete(X_test, 2, 1)
X_test=np.delete(X_test, 2, 1)
```

```python
pd.DataFrame(X_train)
```

```python
pd.DataFrame(X_test)
```

```python
# For Training
import re
import nltk
from collections import Counter
nltk.download("stopwords")
nltk.download('wordnet')
from nltk.corpus import stopwords
all_stopwords=stopwords.words("english")
#from nltk.stem.porter import PorterStemmer
# OR
from nltk.stem import WordNetLemmatizer
for i in range(len(X_train)):
    sentence=re.sub("[^a-zA-Z]", " ", X_train[i][0])
    sentence=sentence.lower()
    sentence=sentence.split()
    #sentence=list(set(sentence))
    #ps=PorterStemmer()
    lemmatizer = WordNetLemmatizer()
    #review=[ps.stem(word) for word in review if not word in set(all_stopwords)]
    sentence=[lemmatizer.lemmatize(word) for word in sentence if not word in set(all_stopwords)]
    counter=Counter(sentence)
    sentence=counter.most_common(10)
    sentence=[word[0] for word in sentence]
    sentence=" ".join(sentence)
    X_train[i][0]=sentence

#For Testing
for i in range(len(X_test)):
    sentence=re.sub("[^a-zA-Z]", " ", X_test[i][1])
    sentence=sentence.lower()
    sentence=sentence.split()
    #sentence=list(set(sentence))
    #ps=PorterStemmer()
    lemmatizer = WordNetLemmatizer()
    #review=[ps.stem(word) for word in review if not word in set(all_stopwords)]
    sentence=[lemmatizer.lemmatize(word) for word in sentence if not word in set(all_stopwords)]
    counter=Counter(sentence)
    sentence=counter.most_common(10)
    sentence=[word[0] for word in sentence]
    sentence=" ".join(sentence)
    X_test[i][1]=sentence
```

## Handling Categorical Data

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(np.append(X_train[:, -1:], X_test[:, -1:], axis=0))
X_train[:, -1]=le.transform(X_train[:, -1:])
X_test[:, -1]=le.transform(X_test[:, -1:])
```

## Creating Bag of Words Model

```python
# from sklearn.feature_extraction.text import CountVectorizer
# cv=CountVectorizer(max_features=1000)
# x_train=cv.fit_transform(X_train[:, 0]).toarray()
# x_test=cv.transform(X_test[:, 1]).toarray()
from sklearn.feature_extraction.text import HashingVectorizer
hv=HashingVectorizer(alternate_sign=False, n_features=1500)
s=10000
hv.partial_fit(X_train[:s, 0])
for i in range(1, (n//s)+1):
    if i!=n//s:
        hv.partial_fit(X_train[s*i:s*(i+1), 0])
    else:
        hv.partial_fit(X_train[s*i:, 0])
```

## Training and Testing the Model

```python
from sklearn.naive_bayes import MultinomialNB
classifier=MultinomialNB()

for i in range(2):
    s=25000
    if i==0:
        classifier.partial_fit(np.concatenate((hv.transform(X_train[:s, 0]).toarray(), X_train[:s, -1:]), axis=1), y_train[:s], classes=list(set(y_train)))
    else:
        classifier.partial_fit(np.concatenate((hv.transform(X_train[:s, 0]).toarray(), X_train[:s, -1:]), axis=1), y_train[:s])
    for x in range(1, (n//s)+1):
        print("#", end="")
        if x!=n//s:
            classifier.partial_fit(np.concatenate((hv.transform(X_train[s*x:s*(x+1), 0]).toarray(), X_train[s*x:s*(x+1), -1:]), axis=1), y_train[s*x:s*(x+1)])
        else:
            classifier.partial_fit(np.concatenate((hv.transform(X_train[s*x:, 0]).toarray(), X_train[s*x:, -1:]), axis=1), y_train[s*x:])
    print("Training-"+str(i+1)+" Done")

s=500
y_pred=classifier.predict(np.concatenate((hv.transform(X_test[:s, 1]).toarray(), X_test[:s, -1:]), axis=1))
for x in range(1, 110775//s):
    y_pred=np.append(y_pred, classifier.predict(np.concatenate((hv.transform(X_test[s*x:s*(x+1), 1]).toarray(), X_test[s*x:s*(x+1), -1:]), axis=1)))
y_pred=np.append(y_pred, classifier.predict(np.concatenate((hv.transform(X_test[110500:, 1]).toarray(), X_test[110500:, -1:]), axis=1))) 
print("Prediction Completed")
```
